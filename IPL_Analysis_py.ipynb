```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

matches = pd.read_csv("/content/matches.csv")
deliveries = pd.read_csv("/content/delivery.csv")

print(matches.head())
print(matches.info())
print(deliveries.head())



```

```python
team_wins = matches['winner'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=team_wins.values, y=team_wins.index, palette="viridis")
plt.title("üèÜ Top 10 Teams by Total Wins in IPL")
plt.xlabel("Number of Wins")
plt.ylabel("Teams")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='toss_decision', hue='result', data=matches)
plt.title("Toss Decision vs Match Result")
plt.xlabel("Toss Decision")
plt.ylabel("Match Count")
plt.show()
```

```python
season_matches = matches['season'].value_counts().sort_index()

plt.figure(figsize=(10,5))
sns.lineplot(x=season_matches.index, y=season_matches.values, marker="o")
plt.title("üìÖ Number of Matches Played Each Season")
plt.xlabel("Season")
plt.ylabel("Match Count")
plt.show()
```

```python
top_batsmen = deliveries.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_batsmen.values, y=top_batsmen.index, palette="coolwarm")
plt.title("üî• Top 10 Run Scorers in IPL")
plt.xlabel("Total Runs")
plt.ylabel("Batsmen")
plt.show()
```

```python
mom = matches['player_of_match'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=mom.values, y=mom.index, palette="mako")
plt.title("‚≠ê Top 10 Players with Most MOM Awards")
plt.xlabel("Awards")
plt.ylabel("Players")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='season', hue='winner', data=matches)
plt.title("üèÜ Winning Teams Across Seasons")
plt.xlabel("Season")
plt.ylabel("Number of Wins")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
