```python
# ===============================================================
# üé¨ AMAZON PRIME VIDEO - EXPLORATORY DATA ANALYSIS (EDA)
# Author: Omkar Panditkar
# ===============================================================

# üì¶ Importing Required Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 5)

# ===============================================================
# üì• Step 1: Load the Dataset
# ===============================================================

# ‚úÖ If you're using Google Colab, first upload your files:
# from google.colab import files
# uploaded = files.upload()

# Then read:
titles = pd.read_csv("titles.csv")
credits = pd.read_csv("credits.csv")

# ===============================================================
# üîç Step 2: Understanding the Data
# ===============================================================

print("üé¨ Titles Dataset Shape:", titles.shape)
print("üé≠ Credits Dataset Shape:", credits.shape)

print("\nüìò Titles Columns:\n", titles.columns)
print("\nüìò Credits Columns:\n", credits.columns)

print("\nüîπ Titles Data Info:")
print(titles.info())

print("\nüîπ Missing Values in Titles:")
print(titles.isnull().sum())

# ===============================================================
# üßπ Step 3: Data Cleaning
# ===============================================================

# Drop duplicates
titles.drop_duplicates(inplace=True)

# Fill missing numeric values with mean
titles['imdb_score'].fillna(titles['imdb_score'].mean(), inplace=True)
titles['tmdb_popularity'].fillna(titles['tmdb_popularity'].mean(), inplace=True)

# Fill missing text values with 'Unknown'
for col in ['age_certification', 'genres', 'production_countries']:
    titles[col].fillna('Unknown', inplace=True)

# ===============================================================
# üìä Step 4: Exploratory Data Analysis
# ===============================================================

# 4.1 Distribution of show types (Movies vs TV Shows)
plt.figure(figsize=(6,4))
sns.countplot(data=titles, x='show_type', palette='viridis')
plt.title('Distribution of Show Types on Amazon Prime')
plt.xlabel('Show Type')
plt.ylabel('Count')
plt.show()

# 4.2 Top 10 Genres
plt.figure(figsize=(10,5))
all_genres = titles['genres'].dropna().str.split(',').explode().str.strip()
top_genres = all_genres.value_counts().head(10)
sns.barplot(x=top_genres.values, y=top_genres.index, palette='mako')
plt.title('Top 10 Genres on Amazon Prime')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()

# 4.3 IMDb Score Distribution
plt.figure(figsize=(8,5))
sns.histplot(titles['imdb_score'], kde=True, bins=20, color='teal')
plt.title('Distribution of IMDb Scores')
plt.xlabel('IMDb Score')
plt.ylabel('Frequency')
plt.show()

# 4.4 Content released over years
content_per_year = titles['release_year'].value_counts().sort_index()
plt.figure(figsize=(10,5))
sns.lineplot(x=content_per_year.index, y=content_per_year.values, color='coral')
plt.title('Trend of Content Released Over Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.show()

# 4.5 IMDb Score vs TMDB Popularity
plt.figure(figsize=(8,5))
sns.scatterplot(data=titles, x='imdb_score', y='tmdb_popularity', hue='show_type', alpha=0.7)
plt.title('IMDb Score vs TMDB Popularity')
plt.xlabel('IMDb Score')
plt.ylabel('TMDB Popularity')
plt.show()

# 4.6 Top 10 Production Countries
plt.figure(figsize=(10,5))
all_countries = titles['production_countries'].dropna().str.split(',').explode().str.strip()
top_countries = all_countries.value_counts().head(10)
sns.barplot(x=top_countries.values, y=top_countries.index, palette='cool')
plt.title('Top 10 Production Countries on Amazon Prime')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.show()

# ===============================================================
# üìà Step 5: Insights & Observations
# ===============================================================

print("üîπ INSIGHTS & CONCLUSIONS:")
print("1Ô∏è‚É£ Amazon Prime‚Äôs library contains both Movies and TV Shows, with Movies dominating the platform.")
print("2Ô∏è‚É£ Drama, Comedy, and Action are the most frequent genres.")
print("3Ô∏è‚É£ Content production has significantly increased after 2015.")
print("4Ô∏è‚É£ IMDb scores mostly range between 6‚Äì8, showing average to good ratings.")
print("5Ô∏è‚É£ The USA leads in content production, followed by India and the UK.")
print("6Ô∏è‚É£ IMDb score and TMDB popularity have a mild positive relationship.")
print("7Ô∏è‚É£ These insights can help Amazon Prime tailor its content strategy based on audience interests.")

# ===============================================================
# üèÅ Step 6: Save Cleaned Data and Visualizations
# ===============================================================

titles.to_csv("cleaned_titles.csv", index=False)
print("\n‚úÖ Cleaned dataset saved as 'cleaned_titles.csv'.")

# ===============================================================
# üìö End of Project
# ===============================================================

print("\nüéâ EDA Completed Successfully by Omkar Panditkar!")
```
